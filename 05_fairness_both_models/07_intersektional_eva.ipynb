```python
def get_metrics(first, second, firstString, secondString):
    df = pd.concat([first, second])
    df = df.drop(columns=['Unnamed: 0', 'Accuracy'])
    df = pd.pivot_table(df, values=["Precision","Recall","AUC","FPR"], index=["model"], columns=["group"])
    df['PP'] = df.Precision[firstString]-df.Precision[secondString]
    df['EO'] = df.Recall[secondString]-df.Recall[firstString]
    df['SA'] = df.AUC[firstString]-df.AUC[secondString]
    df['PE'] = df.FPR[secondString]-df.FPR[firstString]
    df = df.drop(columns=['AUC','Precision','Recall','FPR'])
    df.columns = df.columns.droplevel(1)
    df = pd.pivot_table(df, values=["PP","EO","SA","PE"],  columns=["model"])
    return df
```

```python
firstString = "df_abi_deutsch"
secondString = "df_abi_migration"
df_abi_erstsprache = get_metrics(df_abi_deutsch,df_abi_migration, firstString, secondString)

firstString = "df_keinAbi_deutsch"
secondString = "df_keinAbi_migration"
df_keinAbi_erstsprache = get_metrics(df_keinAbi_deutsch,df_keinAbi_migration, firstString, secondString)

firstString = "df_abi_w"
secondString = "df_abi_m"
df_abi_gender = get_metrics(df_abi_w, df_abi_m, firstString, secondString)

firstString = "df_keinAbi_w"
secondString = "df_keinAbi_m"
df_keinAbi_gender = get_metrics(df_keinAbi_w,df_keinAbi_m, firstString, secondString)

firstString = "df_abi_buch1"
secondString = "df_abi_buch0"
df_abi_buch = get_metrics(df_abi_buch1, df_abi_buch0, firstString, secondString)

firstString = "df_keinAbi_buch1"
secondString = "df_keinAbi_buch0"
df_keinAbi_buch = get_metrics(df_keinAbi_buch1,df_keinAbi_buch0, firstString, secondString)

firstString = "df_w_deutsch"
secondString = "df_w_migration"
df_w_erstsprache = get_metrics(df_w_deutsch, df_w_migration, firstString, secondString)

firstString = "df_m_deutsch"
secondString = "df_m_migration"
df_m_erstsprache = get_metrics(df_m_deutsch,df_m_migration, firstString, secondString)

firstString = "df_w_buch1"
secondString = "df_w_buch0"
df_w_buch = get_metrics(df_w_buch1, df_w_buch0, firstString, secondString)

firstString = "df_m_buch1"
secondString = "df_m_buch0"
df_m_buch = get_metrics(df_m_buch1,df_m_buch0, firstString, secondString)


firstString = "df_migration_buch1"
secondString = "df_migration_buch0"
df_migration_buch = get_metrics(df_migration_buch1, df_migration_buch0, firstString, secondString)

firstString = "df_deutsch_buch1"
secondString = "df_deutsch_buch0"
df_deutsch_buch = get_metrics(df_deutsch_buch1,df_deutsch_buch0, firstString, secondString)
```

```python
#format result

def threshold001(v, props=''):
    return props if (v > 0.02) or (v < -0.02) else None

def threshold005(v, props=''):
    return props if (v > 0.05) or (v < -0.05) else None

def negativeValue(v, props=''):
    return props if (v < 0) else None

def showTable(df):
    styled = df.style.set_properties(color="black", align="right")\
        .set_properties(**{'background-color': 'white'})\
        .applymap(threshold001, props='color:orange;')\
        .applymap(threshold005, props='color:red;')\
        .applymap(negativeValue, props='font-weight:bold;')

    return styled
```
